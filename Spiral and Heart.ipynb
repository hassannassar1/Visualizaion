```python
# Load Libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib auto
```

<h3> Draw Spiral </h3>

```python
n = 600 #this control density of points

angle = np.linspace( 1, 10*np.pi, n )
radius = np.linspace( .1, 5, n )

x = radius * np.cos( angle )
y = radius * np.sin( angle )

c, = plt.plot( x[0], y[0], 'ro-')
plt.axis([-5,5,-5,5])
for i in range(n):
    plt.setp(c,data=[x[:i],y[:i]])
    #c, = plt.plot( x[i], y[i], marker='o',\markersize=5, linewidth=.25, c='#559977' )
    plt.pause( .0000000001 )
```

<h3> Draw Heart </h3>

```python
%matplotlib inline
x = np.linspace( -2, 2, 200 )
r = 2
y = np.sqrt( ( r**2 ) - ( x**2 ) )
# I Used Weights and the parameters of plot to tune the heart  
w = [ 1.25, 0.75 ]
plt.fill( w[0] * -y, ( y - w[1] * x ), c='r')
plt.fill( w[0] * y, ( y - w[1] * x ), c='r' )
plt.title("Heart using parameter tuning of circle")
plt.axis('equal')
plt.show()
```

```python
# I found that code on:
# https://www.codesansar.com/python-programming-examples/plot-heart-shape-using-numpy-matplotlib.htm

theta = np.linspace(0, 2 * np.pi, 100)

x = 16 * ( np.sin(theta) ** 3 )
y = 13 * np.cos(theta) - 5* np.cos(2*theta) - 2 * np.cos(3*theta) - np.cos(4*theta)

plt.fill(x, y,'r')
plt.title('Heart Shape')
plt.show()
```
