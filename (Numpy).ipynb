<h5> Import Libraries</h5>

```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

<h5> Ask User for start, step and stop</h5>

```python
x = input("Input Start, Stop and Step: ").split(' ')
```

```python
X = np.linspace(int(x[0]),int(x[1]),int(x[2]))
```

<h5>Ask User for equation</h5>

```python
#np.exp(X)*np.cos(X)
eq = input("Enter Equation in Numpy Format: ")
```

```python
Y = eval(eq)
```

<h5> Plot the equation</h5>

```python
plt.plot(X,Y)
plt.show()
```

<h5> determine X-Crossing</h5>

```python
o = np.append(Y[1:],0)
Y1 = Y*o
y_zeroz = (Y[Y1<=0])[:-1]
x_zeroz = (X[Y1<=0])[:-1]
y_zeroz
```

<h5>Plot X-Crossing</h5>

```python
plt.plot(X,Y)
plt.plot(x_zeroz,y_zeroz,marker = 'x',color='r',linestyle='')
plt.show()
```

```python
x_sub = (X - np.append(X[1:X.shape[0]],0))
y_sub = (Y - np.append(Y[1:Y.shape[0]],0))

dev = y_sub / x_sub
dev = (dev * np.append(dev[1:dev.shape[0]],0))

integ = integrate.cumtrapz(X,Y)

plt.plot(X,Y)
plt.plot(x_zeroz,y_zeroz,marker = 'x',color='r',linestyle='')
plt.plot(X,dev)
plt.plot(X[0:-1],integ)
plt.plot((X[abs(dev)<0.0001])[0:-1],(Y[abs(dev)<0.0001])[0:-1], marker='o',linestyle='')
plt.legend(["X vs Y","Zero Points",'Y/X','Integration','MinMax'])
plt.show() 
```
