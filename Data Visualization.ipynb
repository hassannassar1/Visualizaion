```python
import numpy as np
import plotly.express as px
import pandas as pd

import dash
from dash import html, dcc
import plotly.express as px
import pandas as pd
from dash.dependencies import Input,Output,State
```

```python
df = pd.read_csv(r"COVID-19 Dataset.csv")
t = pd.read_csv(r"time_series.csv",index_col="Country/Region")
t.head()
```

```python
t.drop(['Province/State', 'Lat', 'Long'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
for x in df.columns:
    if df[x].dtype != "object":
        df[x] = df[x].fillna(df[x].mean())  
```

```python
df.dtypes
```

```python
df = df.dropna(axis = 0)
```

```python
df['Last_Updated_Date'] = pd.to_datetime(df['Last_Updated_Date'])
df['day'] = df['Last_Updated_Date'].dt.day
df['month'] = df['Last_Updated_Date'].dt.month
df['year'] = df['Last_Updated_Date'].dt.year
df.drop('Last_Updated_Date',axis = 1 , inplace = True)
```

```python
# lunching App
my_css_link='https://codepen.io/chriddyp/pen/bWLwgP.css'
app =dash.Dash(external_stylesheets=[my_css_link])
```

```python
fig1= px.scatter(df[df.year == min(df.year)],x='Total_Cases',y='Total_Deaths',color='Continent',size='Total_tests',size_max=60,hover_name='Country',
           title='Covid 19',template='plotly_dark',animation_frame='year')
fig2 = px.sunburst(df,color='Country',values='Total_Cases',path=['Continent','Country'],hover_name='Country')

fig3 = px.histogram(df,x='Continent',y='Total_Cases',color='Continent',
            hover_name='Continent',barmode='group')
#fig4 = px.imshow(df[df.columns[:-10]].corr())
fig4 = px.line(t, x=t.index, y = t.columns[1:])
```

```python
app.layout = html.Div([
    html.H1("Covid 19 Data Visualization",style={'textAlign' : 'center','color':'Black' }),
    html.Div([
    html.Div([
            html.Div([
               
            html.H3("Scatter Visualization for Continents",style={'textAlign' : 'center','color':'Black' }),
        html.Div([
            html.Div([
                "Region",
                dcc.Dropdown(
                    id = 'my_dropdown_0',
                    options=[
                        {'label': str(Continent),
                         'value': str(Continent)}
                    for Continent in df.Continent.unique()
                    ],
                    value=df['Continent']),
            ],className='four columns'),
            
            html.Div([
                "X Label",
                dcc.Dropdown(
                    id = 'my_dropdown_1',
                    options=[
                        {'label': str(y),
                         'value': str(y)}
                    for y in df.columns[2:]
                    ],
                    value='Total_Cases'),
            ],className='four columns')],className='row'),
            html.Div([
                html.Div([
                "Y Label",
                dcc.Dropdown(
                    id = 'my_dropdown_2',
                    options=[
                    {'label': str(y),
                     'value': str(y)}
                    for y in df.columns[2:]
                    ],
                    value='Total_Deaths'),
            ],className='four columns'),
                html.Div([ 
                    html.Button('Submit', id='my_button_0', n_clicks=0),
                ],className='four columns')
            
                
            ],className='row'),
            dcc.Graph(id='my_graph1',figure= fig1),
            
           
            
        ],className='six columns'),
        
        
    ]),
    html.Div([
        
        html.H3("Bar Chart For Continents",style={'textAlign' : 'center','color':'Black' }),
        html.Div([
            html.Div([
                "Aggregated Column",
                dcc.Dropdown(
                    id = 'my_dropdown_3',
                    options=[
                        {'label': str(y),
                         'value': str(y)}
                    for y in df.columns[2:]
                    ],
                    value='Total_Cases'),
              ]),
            html.Button('Submit', id='my_button_1', n_clicks=0),
            dcc.Graph(id='my_graph2',figure= fig3),
            
            
        ],style={"margin-left": "0px","margin-right": "50px"},
            className = 'six columns')
    ])], className = 'row'),
    
    html.Div([
        html.Div([
            html.H3("Time Series ",style={'textAlign' : 'center','color':'Black' }),
            dcc.Graph(figure= fig4)
        ],className='six columns'),
        
        html.Div([
            html.H3("Percentages Of Covid Cases In Continents",style={'textAlign' : 'center','color':'Black' }),
            dcc.Graph(figure= fig2)
        ],style={"margin-left": "0px"},className='six columns'),
        
    ],style={"margin-left": "0px","margin-right": "25px"}, className = 'row')              
                      
])
```

```python
@app.callback(
    Output('my_graph1','figure'),
    State('my_dropdown_0','value'),
    State('my_dropdown_1','value'),
    State('my_dropdown_2','value'),
    Input('my_button_0','n_clicks')
    )

def callback(input_1,input_2,input_3,input_4):
    
    if input_1 == None:
        f1 = px.scatter(df,x=input_2,y=input_3,color='Continent',size=input_3,size_max=60,
                        hover_name='Country',animation_frame='year')
    else :
        f1 = px.scatter(df[(df.Continent == input_1)],x=input_2,y=input_3,size=input_3,color='Continent',
                        size_max=60,hover_name='Country',animation_frame='year')
        
    return f1

@app.callback(
    Output('my_graph2','figure'),
    State('my_dropdown_3','value'),
    Input('my_button_1','n_clicks')
    )
def callback1(input_1,input_2):
    f2 =  px.histogram(df,x='Continent',y=input_1,color='Continent',hover_name='Continent',barmode='group')
    return f2  
    
```

```python
app.run_server()
```
