```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
theta = np.linspace(0, 2*np.pi, 200)
r = 2
x1 = r*np.cos(theta)
x2 = r*np.sin(theta)
x11 = x1+3
x22 = x2+3
fig, ax = plt.subplots(1)
ax.fill(x1,x22, c='y' )
ax.fill(x11,x2, c='y' )
ax.fill(x1-3,x2, c='y' )
ax.fill(x1,x2-3, c='y' )

r = 2.5
x1 = r*np.cos(theta)
x2 = r*np.sin(theta)
ax.fill(x1, x2,c='r')

plt.axis('equal')
#plt.xlim(-2,2)
#plt.ylim(-2,2)
plt.show()
```
